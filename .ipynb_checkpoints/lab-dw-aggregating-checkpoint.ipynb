1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
# Your code goes here
import pandas as pd
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv'

# Load dataset
df = pd.read_csv(url)
df

```

```python
#Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign

df2=df[(df['Total Claim Amount']>1000) & (df['Response']=='Yes')]
df2


```

```python
df.columns
```

```python
#Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

df3=df[df['Response']=='Yes']

# Create the pivot table
pivot_table_tca = df3.pivot_table(values='Total Claim Amount', index='Policy Type',columns='Gender', aggfunc='mean')

# Round the pivot table
pivot_table_tca = pivot_table_tca.round(2)

# Show the result
print(pivot_table_tca)

```

```python
#Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

df['State'].value_counts()
```

```python
#Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.
# Groupby data by state to count the number of clients
customers_by_state = df.groupby('State').size().reset_index(name='Number of customers')

# Filter the states with more than 500 customers
states_500_customers = customers_by_state[customers_by_state['Number of customers'] > 500]

# Show the results
print(states_500_customers)
```

```python
#Find the maximum customer lifetime value by education level and gender. Write your conclusions.

# Create the pivot table
pivot_table_maximum_clv = df.pivot_table(values='Customer Lifetime Value', index='Education', columns='Gender', aggfunc='max')

# Round the pivot table
pivot_table_maximum_clv = pivot_table_maximum_clv.round(2)

# Show the result
print(pivot_table_maximum_clv)
```

```python
#Find the minimum customer lifetime value by education level and gender. Write your conclusions.

# Create the pivot table
pivot_table_minimum_clv = df.pivot_table(values='Customer Lifetime Value', index='Education', columns='Gender', aggfunc='min')

# Round the pivot table
pivot_table_minimum_clv = pivot_table_minimum_clv.round(2)

# Show the result
print(pivot_table_minimum_clv)
```

```python
#Find the median customer lifetime value by education level and gender. Write your conclusions.
# Create the pivot table
pivot_table_mean_clv = df.pivot_table(values='Customer Lifetime Value', index='Education', columns='Gender', aggfunc='mean')

# Round the pivot table
pivot_table_mean_clv = pivot_table_mean_clv.round(2)

# Show the result
print(pivot_table_mean_clv)
```

```python
# The marketing team wants to analyze the number of policies sold by state and month. Present the data in a table where the months are arranged as columns and the states are arranged as rows.

# Create the pivot table
pivot_table_nop = df.pivot_table(values='Number of Policies', index='State', columns='Months Since Policy Inception', aggfunc='sum')

# Round the pivot table
pivot_table_nop = pivot_table_nop.round(2)

# Show the result
print(pivot_table_nop)
```

```python
#Display a new DataFrame that contains the number of policies sold by month, by state, for the top 3 states with the highest number of policies sold.

# Group by state
policies_by_state = df.groupby(['State'])['Number of Policies'].sum().reset_index()

# Find the top 3 states
top_3_states = policies_by_state.sort_values(by='Number of Policies', ascending=False).head(3)

# Filter the df by the top 3 states
filtered_df = df[df['State'].isin(top_3_states['State'])]

# Group by state and months, then count
policies_by_state_month = filtered_df.groupby(['State', 'Months Since Policy Inception']).size().reset_index(name='Number of policies sold')

# Create the pivot table
pivot_table_top3 = policies_by_state_month.pivot_table(values='Number of policies sold', 
                                                       index='State', 
                                                       columns='Months Since Policy Inception', 
                                                       aggfunc='sum')

# Show the results
print(pivot_table_top3)
```
